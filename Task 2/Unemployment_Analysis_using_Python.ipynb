```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame
sns.countplot(data=df, x='Region')
plt.title('Count of Regions')
plt.xlabel('Region')
plt.ylabel('Count')
plt.show()
```

```python
color = ['Red','Blue','Green','Orange','Purple']
Region = ['South','East','West','North','Northeast']
```

```python
for i in range(3):
    x = df[df['Region'] == Region[i]]
    plt.scatter(x['Estimated Unemployment Rate'], x['Estimated Employed'], c = color[i], label = Region[i])

plt.xlabel("Unemployment Rate")
plt.ylabel("Estumated Emaployed")
plt.legend()
```

```python
for i in range(3):
    x = df[df['Region'] == Region[i]]
    plt.scatter(x['Estimated Employed'], x['Estimated Labour Participation Rate'], c = color[i], label = Region[i])

plt.xlabel("Estimeted Employed")
plt.ylabel("Estimated Labour Participation")
plt.legend()
```
